```python
from pyspark.sql.functions import col, monotonically_increasing_id, max

df_silver = spark.table("silver_titulos")

df_gold = df_silver.select(
    col("empresa"),
    col("nCodTitulo"),

    col("cCPFCNPJCliente"),
    col("cCodCateg"),
    col("cNatureza"),
    col("cNumDocFiscal"),
    col("cNumParcela"),
    col("cStatus"),

    col("dDtEmissao"),
    col("dDtPrevisao"),
    col("dDtVenc"),
    col("dDtPagamento"),

    col("nCodCC"),
    col("nCodCliente"),
    col("nValorTitulo"),

    col("cLiquidado"),
    col("nDesconto"),
    col("nJuros"),
    col("nMulta"),
    col("nValAberto"),
    col("nValLiquido"),
    col("nValPago"),
)
```

```python
display(df_gold)
```

```python
df_gold = df_gold.withColumn("sk_titulo", monotonically_increasing_id())
```

```python
df_gold.write.format("delta").mode("overwrite").saveAsTable("gold_fato_titulos")
```
