```python
from pyspark.sql.functions import col, monotonically_increasing_id

df_silver = spark.table("silver_contas")

df_gold = df_silver.select(
    col("nCodCC"),
    col("descricao"),
    col("codigo_banco"),
    col("codigo_agencia"),
    col("numero_conta_corrente"),
    col("tipo_conta_corrente"),
    col("empresa")
)


df_gold = df_gold.withColumn("sk_conta", monotonically_increasing_id())
```

```python
display(df_gold)
```

```python
df_gold.write.format("delta").mode("overwrite").saveAsTable("gold_dim_contas")
```
