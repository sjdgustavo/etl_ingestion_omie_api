```python
from pyspark.sql.functions import col, monotonically_increasing_id

df_silver = spark.table("silver_categorias")
```

```python
df_silver.dtypes
```

```python
df_gold = df_silver.select(
    col("codigo"),
    col("descricao"),
    col("empresa"),
)

df_gold = df_gold.dropDuplicates(["codigo", "empresa"])

df_gold = df_gold.withColumn("sk_categoria", monotonically_increasing_id())
```

```python
display(df_gold)
```

```python
df_gold.write.format("delta").mode("overwrite").saveAsTable("gold_dim_categorias")
```
