```python
from pyspark.sql.functions import col, monotonically_increasing_id, max

df_silver = spark.table("silver_extratos")

df_gold = df_silver.select(
    col("conta"),
    col("nCodLancamento"),
    col("nCodLancRelac"),
    col("cSituacao"),
    col("dDataLancamento"),
    col("nValorDocumento"),
    col("nSaldo"),
    col("cNatureza"),
    col("empresa"),
    col("tipo_linha")
)
```

```python
df_gold = df_gold.filter(col("tipo_linha") == "movimentacao")
df_gold = df_gold.filter(col("cSituacao").isin("Conciliado", "Não conciliado"))
```

```python
display(df_gold)
```

```python
df_gold.select("cSituacao").distinct().show()
```

```python
df_gold = df_gold.withColumn("sk_extrato", monotonically_increasing_id())
```

```python
df_gold.write.format("delta").mode("overwrite").saveAsTable("gold_fato_extratos")
```
