```python
from pyspark.sql.functions import col, monotonically_increasing_id

df_silver = spark.table("silver_clientes")

df_gold = df_silver.select(
    col("cnpj_cpf"),
    col("codigo_cliente_integracao"),
    col("codigo_cliente_omie"),
    col("empresa"),
    col("inativo_flag"),
    col("nome_fantasia"),
    col("pessoa_fisica"),
    col("razao_social"),

)

df_gold = df_gold.dropDuplicates(["codigo_cliente_omie", "empresa"])

df_gold = df_gold.withColumn("sk_cliente", monotonically_increasing_id())
```

```python
df_gold.write.format("delta").mode("overwrite").saveAsTable("gold_dim_clientes")
```
